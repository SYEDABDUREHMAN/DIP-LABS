```python
import numpy as np
from PIL import Image

# Open the image
img = Image.open("D:\\UNI\\DIP SIR AEHTESHAM\\TIF FILE\\HV-ALPSRP268500490-H2.2_UA.tif")

# Convert the image to a NumPy array
img_array = np.array(img)

# Calculate image statistics
mean = np.mean(img_array)
std_dev = np.std(img_array)
min = np.min(img_array)
max = np.max(img_array)

# Print the results
print("Mean:", mean)
print("Standard Deviation:", std_dev)
print("Min:", min)
print("Max:", max)
```

```python
#Reducing Size of the image
import matplotlib.pyplot as plt
import cv2
import numpy as np

 
try:
    img = cv2.imread("D:\\UNI\\DIP SIR AEHTESHAM\\TIF FILE\\HV-ALPSRP268500490-H2.2_UA.tif") 
    (height, width) = img.shape[:2]    
    res = cv2.resize(img, (int(width / 4), int(height / 4)), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite('result.tiff', res)
    plt.imshow(res)
 
except IOError:
    print('Error while reading files !!!')
```

```python
#Rotating Tiff Images
try:
    img = cv2.imread("D:\\UNI\\DIP SIR AEHTESHAM\\TIF FILE\\HV-ALPSRP268500490-H2.2_UA.tif")
    (rows, cols) = img.shape[:2]
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 180, 1)
    res1 = cv2.warpAffine(img, M, (cols, rows))
    cv2.imwrite('result.tiff', res1)
    plt.imshow(res1)
except IOError:
    print('Error while reading files !!!')
```

```python
#Translating Matrix
M = np.float32([[1, 0, 100], [0, 1, 50]])
try:
    img = cv2.imread("D:\\UNI\\DIP SIR AEHTESHAM\\TIF FILE\\HV-ALPSRP268500490-H2.2_UA.tif")
    res2 = cv2.warpAffine(img, M, (cols, rows))
    cv2.imwrite('result.tiff', res2)
    plt.imshow(res2)
 
except IOError:
    print('Error while reading files !!!')
```

```python
try:
    img = cv2.imread("D:\\UNI\\DIP SIR AEHTESHAM\\TIF FILE\\HV-ALPSRP268500490-H2.2_UA.tif")
    edges = cv2.Canny(img, 15, 10)
    cv2.imwrite('result.tiff', edges)
    plt.imshow(edges)

except IOError:
    print('Error while reading files !!!')
```
